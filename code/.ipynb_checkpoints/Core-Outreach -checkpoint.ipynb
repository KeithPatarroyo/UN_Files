### Initial Database:

http://www.phon.ox.ac.uk/corpus

#### Sample description:

*"English speakers were divided into 2 groups (12 speakers each). The first group read the same 45 texts as the speakers of other languages and retold Cinderella. The second group did not read or retell Cinderella. Instead these speakers repeated 4 texts three times. Each repetition was recorded on a separate day"*

### Procedure

#### Filtering:

- The Corpus is made up of paragraphs from J.K. Rowling. Harry Potter and the Chamber of Secrets. Scholastic Paperbacks, 2000; and from Cinderella (A total of 46 paragraphs)
- Each recording was labeled by "fast", "normal" or "slow" depending on the pace at which the speakers read it. We filtered slow recordings.
- The top 5 slowest recordings of each paragraph were filtered.

#### Search for vowels:

- Using the software Audacity®, the recordings were "cutted" in order to isolate the vowels, and the noise background was cleaned.

An example is presented in the follwing lines

## Example 1

### Words:  <u>A</u>fter, th<u>a</u>t

### Phonetic Transcription in British English:  <u>ˈɑːftə </u>, <u>ˈðæt</u>

### Chosen vowels: <u>A</u> --  <u>ɑ </u>, <u>a</u> --  <u>æ </u>

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot

import numpy as np

import warnings
warnings.filterwarnings('ignore')

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
import matplotlib

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
sound_1 = thinkdsp.read_wave('example1.wav')  # 'After' sound by individual A.
sound_2 = thinkdsp.read_wave('example_2.wav')  # 'After' cut sound by individual A.
sound_3 = thinkdsp.read_wave('example_3.wav')  # 'After' sound by individual B.
sound_4 = thinkdsp.read_wave('example_4.wav')  # 'that' sound by individual B.
```

```python
sound_1.make_audio()
```

```python
sound_2.make_audio()
```

```python
sound_3.make_audio()
```

```python
sound_4.make_audio()
```

## First sound

```python
start = 0.001
duration = 0.9
segment = sound_1.segment(start, duration)
segment.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
spectrum = segment.make_spectrum()
```

```python
spectrum.plot(high=4000)
thinkplot.config(xlabel='Frequency (Hz)')
```
