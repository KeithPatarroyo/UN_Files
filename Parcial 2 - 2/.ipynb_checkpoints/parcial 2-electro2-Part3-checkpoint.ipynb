```python
import cmath 
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#Problem parameters
omega_p = 0.49
d = 100.0
c = 1.0
```

## Part 3

### Transmition coefficcients  $t_{1-4}=t'_{1-4}$ :

```python
def t_1_4(omega_p,d,c,omega):
    num = 2.*np.sqrt(1.-(omega_p/omega)**2)
    A = (np.cos(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*2.*np.sqrt(1.-(omega_p/omega)**2)-1j*np.sin(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*(2.-(omega_p/omega)**2))
    B = (np.cos(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*3.*np.sqrt(1.-(omega_p/omega)**2)-1j*np.sin(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*(3.-(omega_p/omega)**2))
    denom = np.cos((np.sqrt(2)*omega*d)/c)*A-1j*np.sin((np.sqrt(2)*omega*d)/c)*B
    return num/denom
```

```python
omega = np.arange(0.5,1.,0.0005)
```

```python
t_1_4_list = t_1_4(omega_p,d,c,omega)
```

```python
t_modules = []
t_modules_2 = []
t_phases = []
for t_coef in t_1_4_list:
    t_modules.append(cmath.polar(t_coef)[0])
    t_modules_2.append(cmath.polar(t_coef)[0]**2)
    t_phases.append(np.angle(t_coef))
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Modulus $|t\'_{1-4}|$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega,t_modules, 'b-', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Modulus $|t_{1-4}|$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega,t_modules_2, 'b-', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Phase ${Arg}(t\'_{1-4})$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega,t_phases, 'r', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

### Reflection coefficcients  $r'_{1-4}=r_{1-4}$ :

```python
def r_1_4(omega_p,d,c,omega):
    num = (1j*np.sin(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*((1.-np.sqrt(2.))-(omega_p/omega)**2))+np.cos(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*(1.-np.sqrt(2.))*np.sqrt(1.-(omega_p/omega)**2)
    denom = (np.cos(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*(1.+np.sqrt(2.))*np.sqrt(1.-(omega_p/omega)**2)-1j*np.sin(omega*d/(c/np.sqrt(1.-(omega_p/omega)**2)))*((1.+np.sqrt(2.))-(omega_p/omega)**2))
    return num/denom
```

```python
omega_r = np.arange(0.5,1.,0.0005)
```

```python
r_1_4_list = r_1_4(omega_p,d,c,omega_r)
```

```python
r_modules = []
r_phases = []
r_modules_2 = []
for r_coef in r_1_4_list:
    r_modules.append(cmath.polar(r_coef)[0])
    r_modules_2.append(cmath.polar(r_coef)[0]**2)
    r_phases.append(np.angle(r_coef))
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Modulus $|r\'_{1-4}|$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega_r,r_modules, 'b-', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Modulus $|r_{1-4}|$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega_r,r_modules_2, 'b-', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Phase ${Arg}(r\'_{1-4})$',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega_r,r_phases, 'r', linewidth=2.0,label="Glass")
#plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Modulus ',fontsize=20)
plt.xlabel('Frequency  $\omega$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(omega_r,np.array(r_modules_2), 'b-', linewidth=2.0,label="$\mathcal{T}$, Transmitance")
plt.plot(omega_r,(np.sqrt(2))*np.array(t_modules_2), 'r-', linewidth=2.0,label="$\mathcal{R}$, Reflectance")
plt.plot(omega_r,(np.sqrt(2))*np.array(t_modules_2)+np.array(r_modules_2), 'g-', linewidth=2.0,label="$\mathcal{R}+\mathcal{T}=1$")
plt.ylim([0,1.2])
plt.legend(loc='upper right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

### Electric field in the interval $[-d,2d]$

```python
region_4 = np.arange(d,2.*d,d/10000.)
region_2 = np.arange(0.,d,d/10000.)
region_0 = np.arange(-d,0.,d/10000.)
omega_e = 0.5
```

```python
def field_region_4(omega_e,c,region,omega_p,d):
    return np.exp(1j*(omega_e/(c/np.sqrt(2)))*region)*t_1_4(omega_p,d,c,omega_e)
field_4 = field_region_4(omega_e,c,region_4-d,omega_p,d)
field_4_real = np.real(field_4)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.ylabel('Re($\mathbf{E}(z)$)',fontsize=20)
plt.xlabel('Distance  $z$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(region_4,field_4_real, 'b-', linewidth=2.0,label="Region 4")
plt.legend(loc='lower right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
field_2_1 = np.exp(-1j*(omega_e/(c/np.sqrt(1.-(omega_p/omega_e)**2)))*region_2)*((np.sqrt(1.-(omega_p/omega_e)**2)+np.sqrt(2))/(2.*np.sqrt(1.-(omega_p/omega_e)**2)))*field_region_4(omega_e,c,0.,omega_p,d)
field_2_2 = np.exp(1j*(omega_e/(c/np.sqrt(1.-(omega_p/omega_e)**2)))*region_2)*((np.sqrt(1.-(omega_p/omega_e)**2)-np.sqrt(2))/(2.*np.sqrt(1.-(omega_p/omega_e)**2)))*field_region_4(omega_e,c,0.,omega_p,d)
field_2_real = np.real(field_2_1)+np.real(field_2_2)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
width = 12
height = 5
plt.figure(figsize=(width, height))
plt.ylabel('Re($\mathbf{E}(z)$)',fontsize=20)
plt.xlabel('Distance  $z$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(region_2[::-1],field_2_real, 'g-', linewidth=2.0,label=r"Region $\varepsilon(\omega)$")
plt.plot(region_4,field_4_real, 'b-', linewidth=2.0,label=r"Region $2\varepsilon_0$")
plt.legend(loc='upper right', shadow=True,fontsize=15,numpoints=1)
plt.show()
```

```python
field_1_1 = np.exp(-1j*(omega_e/(c/np.sqrt(1.-(omega_p/omega_e)**2)))*d)*((np.sqrt(1.-(omega_p/omega_e)**2)+np.sqrt(2))/(2.*np.sqrt(1.-(omega_p/omega_e)**2)))*field_region_4(omega_e,c,0.,omega_p,d)
field_1_2 = np.exp(1j*(omega_e/(c/np.sqrt(1.-(omega_p/omega_e)**2)))*d)*((np.sqrt(1.-(omega_p/omega_e)**2)-np.sqrt(2))/(2.*np.sqrt(1.-(omega_p/omega_e)**2)))*field_region_4(omega_e,c,0.,omega_p,d)

field_0_1 = np.exp(-1j*(omega_e/c)*region_2)*((field_1_1*(np.sqrt(1.-(omega_p/omega_e)**2)+1.)*0.5)+(field_1_2*(1.-np.sqrt(1.-(omega_p/omega_e)**2))*0.5))
field_0_2 = np.exp(1j*(omega_e/c)*region_2)*((field_1_1*(1.-np.sqrt(1.-(omega_p/omega_e)**2))*0.5)+(field_1_2*(1.+np.sqrt(1.-(omega_p/omega_e)**2))*0.5))

field_0_real = +np.real(field_0_1)+np.real(field_0_2)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
width = 12
height = 5
plt.figure(figsize=(width, height))
plt.ylabel('Re($\mathbf{E}(z)$)',fontsize=20)
plt.xlabel('Distance  $z$(a.u.)',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.plot(region_0[::-1],field_0_real, 'r-', linewidth=2.0,label=r"Region $\varepsilon_0$")
plt.plot(region_2[::-1],field_2_real, 'g-', linewidth=2.0,label=r"Region $\varepsilon(\omega)$")
plt.plot(region_4,field_4_real, 'b-', linewidth=2.0,label=r"Region $2\varepsilon_0$")
plt.legend(loc='upper right', shadow=True,fontsize=15,numpoints=1)
plt.savefig('e_electric_field_test.png', transparent=True)
plt.show()
```

```python
field_0_a_1 = np.exp(-1j*(omega_e/c)*region_2)*1
field_0_a_2 = np.exp(1j*(omega_e/c)*region_2)*1*r_1_4(omega_p,d,c,omega_e)

field_0_a_real = +np.real(field_0_a_1)+np.real(field_0_a_2)
```
